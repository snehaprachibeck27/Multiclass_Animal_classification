```python
# Importing necessary libraries
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image  import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D,  Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report
```

```python
#Path to the dataset
dataset_path =  "/kaggle/input/animal-image-dataset-90-different-animals/animals/animals"

#List all the classes(Animal Types)
classes = os.listdir(dataset_path)
num_classes = len(classes)


#Display the no of claases along with some pattern
print(f'Number of classes:{num_classes}')
print(f'Classes:{classes}')
```

```python
#visulazie some sample images from the dataset
import matplotlib.pyplot as plt
import os
plt.figure(figsize=(15,20))
for i, class_name in enumerate(classes[ :20]):
  class_path = os.path.join(dataset_path, class_name)
  img_name = os.listdir(class_path)[0]
  img_path = os.path.join(class_path, img_name)
  img = plt.imread(img_path)

  plt.subplot(4, 6, i+1)
  plt.imshow(img)
  plt.title(f'{class_name} \n shape: {img.shape}')
  plt.axis('off')
plt.show()
```
